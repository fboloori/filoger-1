```python
import pandas as pd
import numpy as np
import umap
from sklearn.manifold import TSNE
from sklearn.datasets import make_moons
from matplotlib import pyplot as plt
```

```python
X, y = make_moons(n_samples=1000, noise=0.01)
y
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
%%time
tsne = TSNE(n_components=1, perplexity=50 ,random_state=123)
tsne_result = tsne.fit_transform(X)
tsne_result
```

```python
plt.scatter(tsne_result[:, 0], range(0, len(tsne_result)), c=y)
```

```python
import umap
u = umap.UMAP(n_components=1)
u.fit(X)
u_result = u.transform(X)
u_result
```

```python
plt.scatter(u_result[:, 0], range(0, len(u_result)), c=y)
```
