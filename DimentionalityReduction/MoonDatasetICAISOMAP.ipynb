```python
import pandas as pd
import numpy as np
from sklearn.decomposition import FastICA
from sklearn.manifold import Isomap
from sklearn.datasets import make_moons
from matplotlib import pyplot as plt
```

```python
X, y = make_moons(n_samples=1000, noise=0.01)
y
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
ica = FastICA(n_components=1, random_state=123)
ica.fit(X)
ica_result = ica.transform(X)
ica_result
```

```python
plt.scatter(ica_result[:, 0], range(0, len(ica_result)), c=y)
```

```python
%%time
isomap = Isomap(n_components=1)
isomap.fit(X)
isomap_result = isomap.transform(X)
isomap_result
```

```python
plt.scatter(isomap_result[:, 0], range(0, len(isomap_result)), c=y)
```
